```python
b = cosine_similarity(final_user)
np.fill_diagonal(b, 0 )
similarity_with_user = pd.DataFrame(b,index=final_user.index)
similarity_with_user.columns=final_user.index
similarity_with_user
```

```python
cosine = cosine_similarity(final_mynta)
np.fill_diagonal(cosine, 0 )
similarity_with_mynta = pd.DataFrame(cosine,index=final_mynta.index)
similarity_with_mynta.columns=final_user.index
similarity_with_mynta
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
sim_user_30_u = find_n_neighbours(similarity_with_user,11)
sim_user_30_u
```

```python
sim_user_30_m = find_n_neighbours(similarity_with_mynta,11)
sim_user_30_m
```

```python
def get_user_similar_mynta( user1, user2 ):
    common_mynta = Rating_mean[Rating_mean.userId == user1].merge(
    Rating_mean[Rating_mean.userId == user2],
    on = "titleId",
    how = "inner" )
    return common_mynta.merge( mynta, on = 'titleId' )
```

```python
def User_item_score(user,item):
    a = sim_user_30_m[sim_user_30_m.index==user].values
    b = a.squeeze().tolist()
    c = final_mynta.loc[:,item]
    d = c[c.index.isin(b)]
    f = d[d.notnull()]
    mean_user = Mean.loc[Mean['userId'] == user,'rating'].values[0]
    index = f.index.values.squeeze().tolist()
    corr = similarity_with_mynta.loc[user,index]
    fin = pd.concat([f, corr], axis=1)
    fin.columns = ['adg_score','correlation']
    fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
    nume = fin['score'].sum()
    deno = fin['correlation'].sum()
    final_score = mean_user + (nume/deno)
    return final_score
```

```python
Rating_mean = Rating_mean.astype({"titleId": str})
mynta_user = Rating_mean.groupby(by = 'userId')['titleId'].apply(lambda x:','.join(x))
```

```python
def User_item_score1(user, x_output):
    mynta_seen_by_user = check.columns[check[check.index==user].notna().any()].tolist()
    a = sim_user_30_m[sim_user_30_m.index==user].values
    b = a.squeeze().tolist()
    d = mynta_user[mynta_user.index.isin(b)]
    l = ','.join(d.values)
    mynta_seen_by_similar_users = l.split(',')
    mynta_under_consideration = list(set(mynta_seen_by_similar_users)-set(list(map(str, mynta_seen_by_user))))
    mynta_under_consideration = list(map(int, mynta_under_consideration))
    score = []
    for item in mynta_under_consideration:
        c = final_mynta.loc[:,item]
        d = c[c.index.isin(b)]
        f = d[d.notnull()]
        mean_user = Mean.loc[Mean['userId'] == user,'rating'].values[0]
        index = f.index.values.squeeze().tolist()
        corr = similarity_with_mynta.loc[user,index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['adg_score','correlation']
        fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = mean_user + (nume/deno)
        score.append(final_score)
    data = pd.DataFrame({'titleId':mynta_under_consideration,'score':score})
    top_2_recommendation = data.sort_values(by='score',ascending=False).head(2)
    mynta_Name = top_2_recommendation.merge(mynta, how='inner', on='titleId')
    if x_output == 1:
      mynta_Names = mynta_Name.title.values.tolist()
    elif x_output == 2:
      mynta_Names = mynta_Name.brand.values.tolist()
    elif x_output == 3:
      mynta_Names = mynta_Name.price.values.tolist()
    elif x_output == 4:
      mynta_Names = mynta_Name.discount.values.tolist()

    return mynta_Names
```

```python
user = int(input("Enter the user id to recommend : "))
predicted_mynta_title = User_item_score1(user, 1)
predicted_mynta_brand = User_item_score1(user, 2)
predicted_mynta_price = User_item_score1(user, 3)
predicted_mynta_discount = User_item_score1(user, 4)
print(" ")
print("   ")

for i in range(len(predicted_mynta_title)):
    #output = predicted_mynta_title[i] + " | " + predicted_mynta_brand[i] + " | " + predicted_mynta_price[i] + " | " + predicted_mynta_discount[i]
    print(predicted_mynta_title[i], " | ", predicted_mynta_brand[i], " | ", predicted_mynta_price[i], " | ", predicted_mynta_discount[i])
```
